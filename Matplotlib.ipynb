```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
import numpy as np
x=np.linspace(0,5,11)
y=x**2
```

```python
plt.plot(x,y,'r-')
plt.xlabel('X Label')
plt.ylabel('Y Label')
plt.title('Title')
```

```python
#Multiplots
plt.subplot(1,2,1) #subplot(no of rows, no of columns, plot no)
plt.plot(x,y,'r')

plt.subplot(1,2,2)
plt.plot(y,x,'b')
```

```python
#Object oriented method
fig=plt.figure()
axes=fig.add_axes([0.1,0.1,0.8,0.8]) #4 arg: left, bottom, width, height
axes.plot(x,y)
axes.set_xlabel('X Label')
axes.set_ylabel('Y Label')
axes.set_title('Title')
```

```python
fig=plt.figure()
axes1=fig.add_axes([0.1,0.1,0.8,0.8])
axes2=fig.add_axes([0.2,0.5,0.4,0.3])
axes1.plot(x,y)
axes2.plot(y,x)
```

```python
fig,axes=plt.subplots(nrows=3,ncols=3)
#axes.plot(x,y)
plt.tight_layout() #fixes overlaps
```

```python
fig,axes=plt.subplots(nrows=1,ncols=2)
for curr_ax in axes: #axes is a list of axes objects which we had manually created earlier.
  curr_ax.plot(x,y)
```

```python
fig,axes=plt.subplots(nrows=1,ncols=2)
axes[0].plot(x,y)
axes[1].plot(y,x)
```

Figure Size and DPI

```python
fig=plt.figure(figsize=(3,2),dpi=100)
#figsize=(width,height) in inches, dpi: dots per inch
ax=fig.add_axes([0,0,1,1])
ax.plot(x,y)
```

```python
#Save a figure
fig.savefig('my_fig.png')
```

```python
fig=plt.figure()
ax=fig.add_axes([0,0,1,1])
ax.plot(x,x**2,label='x sq')
ax.plot(x,x**3,label='x cubed')

ax.legend(loc=0) #0 is for best location. For other locations refer the jupyter notebook
```

Plot Appearance

```python
fig=plt.figure()
ax=fig.add_axes([0,0,1,1])
ax.plot(x,y,color='black',linewidth=3,alpha=0.5,linestyle='-.',marker='o')# could use RGB Hex Code
# default line width is 1, so this line is thrice as thick
#alpha is to control transparency
#find other options for linestyle

ax.set_xlim([0,1]) #as a list pass lower bound, upper bound
ax.set_ylim([0,2])
```
